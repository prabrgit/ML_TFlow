```python
import os
import random
import shutil

# Define the paths to the three folders
folder1_path = '/content/drive/MyDrive/FWL/FWL'
folder2_path = '/content/drive/MyDrive/OV/OV'
folder3_path = '/content/drive/MyDrive/HWL/HWL'

# Define the path to the training set folder
training_set_path = '/content/drive/MyDrive/waterbot/train'

# Define the path to the test set folder
test_set_path = '/content/drive/MyDrive/waterbot/test'

# Define the percentage of images to select for the training and test sets
training_percentage = 0.5  # 80% for training
test_percentage = 0.5  # 10% for testing

# Create labeled folders inside the training set folder
label1_path = os.path.join(training_set_path, 'FWL')
label2_path = os.path.join(training_set_path, 'OV')
label3_path = os.path.join(training_set_path, 'HWL')
os.makedirs(label1_path, exist_ok=True)
os.makedirs(label2_path, exist_ok=True)
os.makedirs(label3_path, exist_ok=True)

# Create a labeled folder inside the test set folder
test_label1_path = os.path.join(test_set_path, 'FWL')
test_label2_path = os.path.join(test_set_path, 'OV')
test_label3_path = os.path.join(test_set_path, 'HWL')
os.makedirs(test_label1_path, exist_ok=True)
os.makedirs(test_label2_path, exist_ok=True)
os.makedirs(test_label3_path, exist_ok=True)

# Function to randomly select files from a given folder
def select_files(folder_path, num_files):
    all_files = os.listdir(folder_path)
    selected_files = random.sample(all_files, num_files)
    return selected_files

# Calculate the number of files for training and testing from each folder
num_files_folder1 = len(os.listdir(folder1_path))
num_files_folder2 = len(os.listdir(folder2_path))
num_files_folder3 = len(os.listdir(folder3_path))

num_files_train_folder1 = int(num_files_folder1 * training_percentage)
num_files_train_folder2 = int(num_files_folder2 * training_percentage)
num_files_train_folder3 = int(num_files_folder3 * training_percentage)

num_files_test_folder1 = int(num_files_folder1 * test_percentage)
num_files_test_folder2 = int(num_files_folder2 * test_percentage)
num_files_test_folder3 = int(num_files_folder3 * test_percentage)

# Select files for training from each folder
selected_files_train_folder1 = select_files(folder1_path, num_files_train_folder1)
selected_files_train_folder2 = select_files(folder2_path, num_files_train_folder2)
selected_files_train_folder3 = select_files(folder3_path, num_files_train_folder3)

# Move the selected training files to labeled folders in the training set folder
for file_name in selected_files_train_folder1:
    shutil.move(os.path.join(folder1_path, file_name), os.path.join(label1_path, file_name))

for file_name in selected_files_train_folder2:
    shutil.move(os.path.join(folder2_path, file_name), os.path.join(label2_path, file_name))

for file_name in selected_files_train_folder3:
    shutil.move(os.path.join(folder3_path, file_name), os.path.join(label3_path, file_name))

# Select files for testing from each folder
selected_files_test_folder1 = select_files(folder1_path, num_files_test_folder1)
selected_files_test_folder2 = select_files(folder2_path, num_files_test_folder2)
selected_files_test_folder3 = select_files(folder3_path, num_files_test_folder3)

# Move the selected test files to labeled folders in the test set folder
for file_name in selected_files_test_folder1:
    shutil.move(os.path.join(folder1_path, file_name), os.path.join(test_label1_path, file_name))

for file_name in selected_files_test_folder2:
    shutil.move(os.path.join(folder2_path, file_name), os.path.join(test_label2_path, file_name))

for file_name in selected_files_test_folder3:
    shutil.move(os.path.join(folder3_path, file_name), os.path.join(test_label3_path, file_name))
```

```python
import matplotlib.pyplot as plt

# Plot the validation and training data separately
def plot_loss_curves(history):
  """
  Returns separate loss curves for training and validation metrics.
  """
  loss = history.history['loss']
  val_loss = history.history['val_loss']

  accuracy = history.history['accuracy']
  val_accuracy = history.history['val_accuracy']

  epochs = range(len(history.history['loss']))

  # Plot loss
  plt.plot(epochs, loss, label='training_loss')
  plt.plot(epochs, val_loss, label='val_loss')
  plt.title('Loss')
  plt.xlabel('Epochs')
  plt.legend()

  # Plot accuracy
  plt.figure()
  plt.plot(epochs, accuracy, label='training_accuracy')
  plt.plot(epochs, val_accuracy, label='val_accuracy')
  plt.title('Accuracy')
  plt.xlabel('Epochs')
  plt.legend();
```

```python
def compare_historys(original_history, new_history, initial_epochs=5):
    """
    Compares two model history objects.
    """
    # Get original history measurements
    acc = original_history.history["accuracy"]
    loss = original_history.history["loss"]

    print(len(acc))

    val_acc = original_history.history["val_accuracy"]
    val_loss = original_history.history["val_loss"]

    # Combine original history with new history
    total_acc = acc + new_history.history["accuracy"]
    total_loss = loss + new_history.history["loss"]

    total_val_acc = val_acc + new_history.history["val_accuracy"]
    total_val_loss = val_loss + new_history.history["val_loss"]

    print(len(total_acc))
    print(total_acc)

    # Make plots
    plt.figure(figsize=(8, 8))
    plt.subplot(2, 1, 1)
    plt.plot(total_acc, label='Training Accuracy')
    plt.plot(total_val_acc, label='Validation Accuracy')
    plt.plot([initial_epochs-1, initial_epochs-1],
              plt.ylim(), label='Start Fine Tuning') # reshift plot around epochs
    plt.legend(loc='lower right')
    plt.title('Training and Validation Accuracy')

    plt.subplot(2, 1, 2)
    plt.plot(total_loss, label='Training Loss')
    plt.plot(total_val_loss, label='Validation Loss')
    plt.plot([initial_epochs-1, initial_epochs-1],
              plt.ylim(), label='Start Fine Tuning') # reshift plot around epochs
    plt.legend(loc='upper right')
    plt.title('Training and Validation Loss')
    plt.xlabel('epoch')
    plt.show()
```
